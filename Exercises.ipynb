```python
#7) From the titanic data-set what is the average age of the survivors, what was the most
#common Pclass and Sex.
```

```python
titanic = pd.read_csv('titanic_train.csv')
```

```python
titanic.head(2)
```

```python
# Average Age of Survivors

for survivor in titanic.Survived:
    if survivor > 0:
        average = titanic.Age.mean()
        
print("The average age of the survivors is {}".format(average))
        
```

```python
# The Common sex

sex = titanic.Sex
countfemale=0
countmale=0
for x in sex :
    if x =='female':
        countfemale += 1
    elif x =='male':
        countmale += 1
    
     
counting = max(countfemale, countmale)

if counting == countfemale:
     print('The most common sex was Female')
elif counting == countmale:
    print('The most common sex was Male')
    


```
